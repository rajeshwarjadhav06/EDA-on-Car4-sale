## How does the price of cars vary across different manufacturing years?

```python
df['Car_price_L'] = df['Car_price'] / 100000
plt.figure(figsize=(13,5))
sns.barplot(data=df, x='Car_year', y='Car_price_L', hue='Car_year', palette='cividis', legend=False, errorbar=None)
plt.title('Average Car Price by Year')
plt.xlabel('Car Year')
plt.ylabel('Car Price (in Lakhs)')
plt.gca().set_yticklabels([f'{int(tick)}L' for tick in plt.gca().get_yticks()])
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Observation
- Newer cars tend to have higher average prices, with a noticeable upward trend from older to recent years. This reflects depreciation over time and the premium value of newer models. However, certain years may show price spikes due to luxury or high-performance listings.

## Which top 10 Cities are having the maximum car price for sell?

```python
exclude_cities = ['Devanahalli, Bangalore', 'Delhi Cantonment, New Delhi','Sholinganallur, Chennai']

# Step 2: Filter them out from the DataFrame
df_filtered = df[~df['Car_city'].isin(exclude_cities)]
```

```python
# Step 3: Group and scale prices to lakhs using the filtered data
city = df_filtered.groupby('Car_city')['Car_price'].max().sort_values(ascending=False).head(10)
city_lakhs = city / 100000

# Step 4: Plot
plt.figure(figsize=(16, 7))
sns.barplot(x=city_lakhs.index,y=city_lakhs.values,palette='Reds_r',width=0.5)

# Format y-axis
from matplotlib.ticker import FuncFormatter
formatter = FuncFormatter(lambda x, _: f'{int(x)}L')
plt.gca().yaxis.set_major_formatter(formatter)

# Annotate bars
for i, value in enumerate(city_lakhs.values):
    plt.text(i, value + 0.5, f'{value:.1f}L', ha='center', va='bottom', fontsize=10)

# Final touches
plt.xticks(rotation=90)
plt.xlabel("Car City")
plt.ylabel("Max Car Price (in Lakhs)")
plt.title("Top 10 Cities by Maximum Car Price Listed")
plt.tight_layout()
plt.show()
```

## Observation
- The top 10 cities which are having the highest car price for sell are the Banglore, Sampbalpur, Hyderabad and etc.
- The maximum car price for sale and from the top 10 cities is around 9000000 Lakh.

## How does the manufacturing year of a used car affect its resale price?

```python
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import FuncFormatter
avg_price_by_year = (
    df.groupby('Car_year')['Car_price']
    .mean()
    .sort_index()
)


avg_price_by_year_L = avg_price_by_year / 100000
fig, axes = plt.subplots(1, 2, figsize=(16, 5))

sns.barplot(
    x=avg_price_by_year_L.index,
    y=avg_price_by_year_L.values,
    palette='cividis',
    ax=axes[0]
)

axes[0].set_title("Average Car Price by Year (Bar)")
axes[0].set_xlabel("Car Year")
axes[0].set_ylabel("Price (Lakhs)")

# Rotate X-ticks by 90 degrees (BAR PLOT ONLY)
axes[0].tick_params(axis='x', rotation=90)

# Y-axis formatter
axes[0].yaxis.set_major_formatter(
    FuncFormatter(lambda x, _: f'{x:.1f} L')
)
sns.lineplot(
    x=avg_price_by_year_L.index,
    y=avg_price_by_year_L.values,
    marker='o',
    color='green',
    ax=axes[1]
)

axes[1].set_title("Average Car Price by Year (Line)")
axes[1].set_xlabel("Car Year")
axes[1].set_ylabel("Price (Lakhs)")

axes[1].yaxis.set_major_formatter(
    FuncFormatter(lambda x, _: f'{x:.1f} L')
)

axes[1].grid(True)

plt.tight_layout()
plt.show()
```

## Observation


- **Plot 1:** Car Year vs. Price (Scatter + Trendline) There's a clear upward trend: newer used cars tend to
  be priced higher, while older models show lower resale values. The red regression line confirms this
  depreciation pattern, typical in second-hand markets.

- **Plot 2:** Average Price by Year (Lineplot) The average resale price increases steadily with newer
  manufacturing years. This trend reflects how recent models retain more value, while older cars 
  depreciate more sharply. So if just one or two old cars are priced high (like a rare or luxury model), it makes the average look much higher.

## How does ownership type affect car prices in top 10 cities?

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Step 1: Define cities to exclude
exclude_cities = ['Devanahalli, Bangalore', 'Delhi Cantonment, New Delhi', 'Punjabi Bagh','Sholinganallur, Chennai']

# Step 2: Filter the DataFrame
df_filtered = df[~df['Car_city'].isin(exclude_cities)]

# Step 3: Get top 10 cities by max car price (excluding unwanted cities)
top_cities = df_filtered.groupby('Car_city')['Car_price'].max().sort_values(ascending=False).head(10).index

# Step 4: Filter data for those top cities
df_top = df_filtered[df_filtered['Car_city'].isin(top_cities)]

# Step 5: Group by city and ownership type, calculate average price
avg_price_city_owner = df_top.groupby(['Car_city', 'Car_owner'])['Car_price'].mean().reset_index()

# Step 6: Plot
plt.figure(figsize=(13, 6))
sns.barplot(data=avg_price_city_owner, x='Car_city', y='Car_price', hue='Car_owner', palette='cividis')

# Step 7: Final touches
plt.title('Average Car Price by Ownership Type in Top 10 Cities (Filtered)')
plt.xlabel('Car City')
plt.ylabel('Average Price (INR)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
import matplotlib.ticker as ticker

# Step 6: Plot
plt.figure(figsize=(13, 6))
sns.barplot(data=avg_price_city_owner, x='Car_city', y='Car_price', hue='Car_owner', palette='cividis')

# Step 7: Final touches
plt.title('Average Car Price by Ownership Type in Top 10 Cities (Filtered)')
plt.xlabel('Car City')
plt.ylabel('Average Price (INR in Lakhs)')
plt.xticks(rotation=90)

plt.gca().yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:.1f}'))
plt.tight_layout()
plt.show()

```

## Observation

- In the top 10 cities, cars with single owners are usually having higher price than those with second owners. 
- Chennai has a large inventory of premium second-hand cars—brands like BMW, Mercedes-Benz, Audi, and Land Rover are commonly resold here.
