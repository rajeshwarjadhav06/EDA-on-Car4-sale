```python
import pandas as pd
import re

# Step 1: Read the CSV file
df = pd.read_csv(r"Cars_Data.csv")

# Step 2: Define regex extraction functions
def extract_car_name(text):
    match = re.search(r"^(.*?),", str(text))
    return match.group(1) if match else None

def extract_car_year(text):
    match = re.search(r"\b(\d{4})\b", str(text))
    return match.group(1) if match else None

def extract_car_location(text):
    match = re.search(r"Location:\s*(.*?)Rs\.", str(text))
    return match.group(1) if match else None

def extract_car_price(text):
    match = re.search(r"Rs\.\s*([\d,]+)", str(text))
    return match.group(1) if match else None

# Step 3: Apply regex to extract columns
df["car_name"] = df["Containers"].apply(extract_car_name)
df["car_year"] = df["Containers"].apply(extract_car_year)
df["car_location"] = df["Containers"].apply(extract_car_location)
df["car_price"] = df["Containers"].apply(extract_car_price)

# Step 4: Clean price 
df["car_price"] = df["car_price"].str.replace(",", "").astype("Int64")

df.head()
```
