# Importing the necessary Libraries

```python
import requests
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import re
import warnings
warnings.filterwarnings("ignore")
```

# Website Link

```python
url = "https://www.carz4sale.in/used/buy/"
```

```python
response=requests.get(url)
```

```python
response.status_code
```

```python
html_code=response.text
```

```python
html_code[:2000]
```

```python
soup=BeautifulSoup(html_code)
```

```python
soup.text[:2000]
```

## Extracting the data from the single container

```python
container=soup.find('div',attrs={"class":"txtBox"})
container.text
```

# Extracting the data from the single page

```python
container=soup.find_all('div',attrs={"class":"boxRow"})
for c in container:
    print(c.text.strip())
```

```python
len(container)
```

## Scraping the data from the pages which needed

```python
url = "https://www.carz4sale.in/used/buy/?page=1"

for i in range(1,21):
    print(f'https://www.carz4sale.in/used/buy/?page={i}')
```

```python
Title = []

for i in range(1,21):
    url = f'https://www.carz4sale.in/used/buy/?page={i}'

    response = requests.get(url)
    html_code = response.text

    soup = BeautifulSoup(html_code, "html.parser")
    
    continer = soup.find_all('div', attrs={'class':'boxRow'})
    for c in continer:
        Title.append(c.text.strip())
```

```python
len(Title)
```

# Creating the DataFrame

```python
df=pd.DataFrame({"Containers":Title})
df
```

```python
df.head()
```

# Checking For Duplicates

```python
print(df.duplicated().sum())
```

# Converted the file to CSV

```python
df.to_csv("Cars_Data.csv",index = False)
```
